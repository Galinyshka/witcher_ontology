```python
from bs4 import BeautifulSoup

html = '''<div class="pi-data-value pi-font">Heiress to the throne of Cintra<br>Heiress to Inis <a href="/wiki/Ard_Skellig" title="Ard Skellig">Ard Skellig</a> and Inis <a href="/wiki/An_Skellig" title="An Skellig">An Skellig</a><br><a href="/wiki/Princess" class="mw-redirect" title="Princess">Princess</a> of <a href="/wiki/Brugge" title="Brugge">Brugge</a> <small>(claimant)</small><br><a href="/wiki/Duchess" class="mw-redirect" title="Duchess">Duchess</a> of <a href="/wiki/Sodden" title="Sodden">Sodden</a> <small>(claimant)</small><br>Suzerain of <a href="/wiki/Attre" title="Attre">Attre</a> and <a href="/wiki/Abb_Yarra" title="Abb Yarra">Abb Yarra</a> <small>(claimant, given to her <a href="/wiki/False_Ciri" class="mw-redirect" title="False Ciri">double</a>)</small></div>'''
html = '''<div class="pi-item pi-data pi-item-spacing pi-border-color" data-source="abilities">
	
		<h3 class="pi-data-label pi-secondary-font">Abilities</h3>
	
	<div class="pi-data-value pi-font">Superhuman abilities<br>Swordsmanship<br><a href="/wiki/Alchemy" title="Alchemy">Alchemy</a><br><a href="/wiki/Signs" class="mw-redirect" title="Signs">Signs</a></div>
</div>'''
soup = BeautifulSoup(html, "html.parser")
div = soup.find("div", class_="pi-data-value")

# Убираем ненужные теги (small, sup)
for tag in div.find_all(["small", "sup"]):
    tag.decompose()

result = []
current = []

# Обходим всех детей, включая текст и теги
for elem in div.children:
    if getattr(elem, "name", None) == "br":
        # закончили один блок
        text = " ".join(current).strip()
        if text:
            result.append(text)
        current = []
    else:
        # собираем текст текущего блока
        if hasattr(elem, "get_text"):
            current.append(elem.get_text(" ", strip=True))
        elif isinstance(elem, str):
            current.append(elem.strip())

# добавляем последний блок
if current:
    text = " ".join(current).strip()
    if text:
        result.append(text)

# объединяем через запятую
final = ", ".join(result)
print(final)
```

```python
def get_html(page_name:str):
    url = "https://witcher.fandom.com/api.php"
    params = {
        "action": "parse",
        "page": page_name,
        "prop": "text",
        "format": "json"
    }
    r = requests.get(url, params=params)
    data = r.json()
    html = data["parse"]["text"]["*"]
    soup = BeautifulSoup(html, "html.parser")
    return soup
```
